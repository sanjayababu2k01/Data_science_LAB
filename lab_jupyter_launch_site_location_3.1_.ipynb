```python
# Task 2: Mark the success/failed launches for each site on the map

# Initialize the map (optional: you can reuse site_map)
launch_map = folium.Map(location=[28, -80], zoom_start=5)

# Loop through all launches
for index, row in spacex_df.iterrows():
    launch_coordinate = [row['Lat'], row['Long']]
    
    # Check launch outcome
    if row['class'] == 1:
        color = 'green'  # Success
        label = 'Success'
    else:
        color = 'red'    # Failure
        label = 'Failure'
    
    # Add circle for the launch
    folium.Circle(
        launch_coordinate,
        radius=500,
        color=color,
        fill=True
    ).add_child(folium.Popup(f"{row['Launch Site']} : {label}")).add_to(launch_map)
    
# Display the map
launch_map
```

```python
# Initialize the map
site_map = folium.Map(location=[28, -80], zoom_start=5)

# Group by launch site to get total launches and success counts
site_stats = spacex_df.groupby('Launch Site').agg(
    total_launches=('class', 'count'),
    success_count=('class', 'sum')
).reset_index()

# Loop through each launch site
for index, row in site_stats.iterrows():
    # Get site coordinates from launch_sites_df
    site_row = launch_sites_df[launch_sites_df['Launch Site'] == row['Launch Site']].iloc[0]
    site_coordinate = [site_row['Lat'], site_row['Long']]
    
    # Calculate success rate
    success_rate = row['success_count'] / row['total_launches']
    
    # Color based on success rate
    if success_rate == 1.0:
        color = 'green'
    elif success_rate >= 0.5:
        color = 'orange'
    else:
        color = 'red'
    
    # Add circle marker with popup showing site stats
    folium.Circle(
        site_coordinate,
        radius=1000,
        color=color,
        fill=True,
        fill_opacity=0.7
    ).add_child(
        folium.Popup(
            f"{row['Launch Site']}<br>Total Launches: {row['total_launches']}<br>"
            f"Successful Launches: {row['success_count']}<br>"
            f"Success Rate: {success_rate:.2f}"
        )
    ).add_to(site_map)

# Display the map
site_map
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
# Initialize the map
launch_map = folium.Map(location=[28, -80], zoom_start=5)

# Create a MarkerCluster and add it to the map
marker_cluster = MarkerCluster().add_to(launch_map)

# Now you can add individual markers to the cluster
for index, row in spacex_df.iterrows():
    launch_coordinate = [row['Lat'], row['Long']]
    color = 'green' if row['class'] == 1 else 'red'
    label = 'Success' if row['class'] == 1 else 'Failure'
    
    folium.Marker(
        location=launch_coordinate,
        popup=f"{row['Launch Site']} : {label}",
        icon=folium.Icon(color=color)
    ).add_to(marker_cluster)

# Display the map
launch_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from folium.features import DivIcon

# Launch site coordinates
launch_site_coord = [28.563197, -80.576820]  # Example: CCAFS LC-40

# Nearby points with coordinates (replace with actual coordinates you got from map)
nearby_points = {
    "City": [28.3922, -80.6077],
    "Railway": [28.5650, -80.5800],
    "Highway": [28.56789, -80.57712],
    "Coastline": [28.56367, -80.57163]
}

# Loop through each nearby point
for name, coord in nearby_points.items():
    # Calculate distance
    distance_km = calculate_distance(launch_site_coord[0], launch_site_coord[1], coord[0], coord[1])
    
    # Add marker with DivIcon showing distance
    distance_marker = folium.Marker(
        location=coord,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#d35400;"><b>{}: {:10.2f} KM</b></div>'.format(name, distance_km)
        )
    )
    distance_marker.add_to(site_map)
    
    # Draw line connecting launch site and nearby point
    folium.PolyLine(
        locations=[launch_site_coord, coord],
        weight=2,
        color='blue',
        tooltip=f"Distance to {name}: {distance_km:.2f} KM"
    ).add_to(site_map)

# Display updated map
site_map
```
